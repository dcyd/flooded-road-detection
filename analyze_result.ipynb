```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager as fm

file_path = '../perfor_results_modis.xlsx'
data = pd.read_excel(file_path)


# 设置绘图风格
sns.set(style="whitegrid")

# 创建绘图
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Comparison of Performance Metrics between Training Methods', fontsize=16)

# 更改x轴标签
data['typ'] = data['typ'].replace({'mix': 'Mixed Dataset', 'lou': 'Louisiana Dataset'})

# IoU for class_0
sns.boxplot(x='typ', y='class_0_iou', data=data, ax=axes[0, 0])
axes[0, 0].set_ylabel('IoU of Non-flood road')
axes[0, 0].set_xlabel('')

# IoU for class_1
sns.boxplot(x='typ', y='class_1_iou', data=data, ax=axes[0, 1])
axes[0, 1].set_ylabel('IoU of flood road')
axes[0, 1].set_xlabel('')

# F1-score for class_0
sns.boxplot(x='typ', y='class_0_f1', data=data, ax=axes[1, 0])
axes[1, 0].set_ylabel('F1-score of Non-flood road')
axes[1, 0].set_xlabel('')

# F1-score for class_1
sns.boxplot(x='typ', y='class_1_f1', data=data, ax=axes[1, 1])
axes[1, 1].set_ylabel('F1-score of flood road')
axes[1, 1].set_xlabel('')

# 隐藏x轴标签
for ax in axes.flatten():
    ax.set_xlabel('')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=0)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
